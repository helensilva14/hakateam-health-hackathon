```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def fsave(df,df_name):
    writer = pd.ExcelWriter(
        f'{df_name}.xlsx', engine='xlsxwriter')
    df.to_excel(writer,sheet_name='ft')
    writer.save()
    print(f'{df_name}.xlsx has been saved sucessfully.')
    
    

def check_miss(df):
    vars_with_missing = []
    vars_50_plus = []
    for f in df.columns:
        missings = df[f].isnull().sum()
        if missings > 0:
            vars_with_missing.append(f)
            missings_perc = missings/df.shape[0]
            
            if missings_perc > 0.5:
                vars_50_plus.append(f)
                
            #print('Variable {} has {} records ({:.2%}) with missing values'.format(f, missings, missings_perc))
    
    print('In total, there are {} variables with missing values'.format(len(vars_with_missing)))
    print('In total, there are {} variables with more than 50% of missing values'.format(len(vars_50_plus)))
    return vars_50_plus

def imp(df):
    print(f'Importing {df} dataframe')
    df = pd.read_excel(df)
    print(f'Imported {df} dataframe')
    return df

def check_distribution(df):
    return ('\n',df.info(),'\n-Skewness:\n',df.skew(),'\n-Kurtosis\n',df.kurtosis())
```

```python
data = pd.read_csv('hakadata2.csv')
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

def encoding(df_e):
    '''Encoding Categorical Variables as Numerical ones'''
    dfa = df_e.copy()
    le = LabelEncoder()
    var_to_encode = ['Health Insurance - Plan','Gender']
    for col in var_to_encode:
        dfa[col] = le.fit_transform(df_e[col])
        
    return dfa

#def acess_patient_info(ID):
    #criar banco de dados para acessar dados do paciente e como imputar o nome do paciente
 #   '''Acessing patients information on the database'''
  #  pat_info = encoding(>>>>db<<<<['Health Insurance - Plan','Latitude','Longitude','Patient Age','Gender'])
        
   # return pat_info

def scale(df):
    df_s = encoding(df[['Health Insurance - Plan','Latitude','Longitude','Patient Age','Gender']].copy())
    for a in df_s.columns:
         df_s[a] = df_s[a]-np.min(df_s[a])/np.max(df_s[a])-np.min(df_s[a])
    return df_s
    
def calc_knn(df_c):
    '''Calculating 7-NN and predicting the'''
    to_pred = ['Health Insurance - Plan','Latitude','Longitude','Patient Age','Gender']
    doctor = 'Doctor'
    
    dff = df_c.dropna()     
    dfx = scale(dff)
    #dfx = pd.DataFrame(
    #    scaler.fit_transform(
    #        encoding(dff[['Health Insurance - Plan','Latitude',
    #                       'Longitude','Patient Age','Gender']])),columns =['Health Insurance - Plan',
    #                                                                        'Latitude','Longitude','Patient Age','Gender'])
    #dff.loc[:,:] = dff[['Health Insurance - Plan','Latitude','Longitude','Patient Age','Gender']]
    
    dfy = dff[doctor]
    
    knn = KNeighborsClassifier(n_neighbors=7)
    knn.fit(dfx,dfy)    

    #as a prediction we gotta produce a rank of the most "preferred of the neighbors" doctors.
    
    d= knn.predict(encoding(data[['Health Insurance - Plan','Latitude','Longitude','Patient Age','Gender']]))
    
    return d
```
